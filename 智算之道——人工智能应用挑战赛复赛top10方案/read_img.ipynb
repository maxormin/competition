```python
def get_all_data(img_road,label_roads,num,count):
    label = {'sunny':0,'cloudy':1,'others':2}
    with open(label_roads) as f:
        d1 = json.load(f)
    roads = glob.glob(img_road)
    for road in tqdm(roads):
        name = road.split('/')[-1]
        leibie = label[d1[name]]
        img = cv2.imread(road)
        if len(img.shape) != 3:
            continue
        if count % 5==0:
            path = '/home/kesci/work/validata/'+str(leibie)+'/'+str(num[leibie])+'.jpg'
        else:
            path = '/home/kesci/work/data/'+str(leibie)+'/'+str(num[leibie])+'.jpg'
        cv2.imwrite(path,img)
        num[leibie] += 1
        count += 1
    return num,count
```

```python
def get_train():
    num = [0,0,0]
    count = 0
    for i in range(1,5):
        img_roads = '/home/kesci/data1/competition/train_set/Part{}/*.jpg'.format(i)
        label_roads = '/home/kesci/data1/competition/train_set/Part{}.json'.format(i)
        num,count = get_all_data(img_roads,label_roads,num,count)
    # num = np.array(num)
    print('read img train end')
```

```python
def get_test_data(img_road):
    roads = glob.glob(img_road)
    for road in tqdm(roads):
        name = road.split('/')[-1]
        img = cv2.imread(road)
        path = '/home/kesci/work/test/test_1/'+name
        cv2.imwrite(path,img)
```

```python
def get_test():
    img_roads = '/home/kesci/data1/competition/test_set/*.jpg'
    get_test_data(img_roads)
    print('read img train end')
```
